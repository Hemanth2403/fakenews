```python
data_merge = pd.concat([data_fake, data_true], axis = 0)
data_merge.head(10)
```

```python
data = data_merge.drop(['title', 'subject', 'date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace  =True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
  text = text.lower()
  text = re.sub('\[.*?\]', '', text)
  text = re.sub("\\W", " ", text)
  text = re.sub('https?://\S+|www\.\S+', '', text)
  text = re.sub('<.*?>+', '',text)
  text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
  text = re.sub('\n', '', text)
  text = re.sub('\w*\d\w*', '', text)
  return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization = TfidfVectorizer(
    max_features=5000,       # Limit the number of features to the top 5000
    stop_words='english',    # Remove common stopwords
    ngram_range=(1, 2)       # Include unigrams and bigrams
)
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
import joblib
joblib.dump(vectorization, 'vectorizer1.pkl')
```

```python
vectorization = joblib.load('vectorizer1.pkl')
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

NB = MultinomialNB()
NB.fit(xv_train, y_train)
```

```python
pred_nb = NB.predict(xv_test)
```

```python
print("Naive Bayes Accuracy:", NB.score(xv_test, y_test))
```

```python
print("\nClassification Report:\n", classification_report(y_test, pred_nb))
```

```python
def output_label(n):
  if n == 0:
    return "Fake News"
  elif n == 1:
    return "Real News"

def manual_testing(news):
  testing_news = {"text" : [news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordopt)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorization.transform(new_x_test)
  pred_NB = NB.predict(new_xv_test)
  
  return print("NB Prediction: {}".format(output_label(pred_NB[0])))
```
